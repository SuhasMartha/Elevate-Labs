```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("train.csv")
```

```python
# Dataset Overview
print(df.info())
print(df.describe(include='all'))

# Value counts
print("Survived:\n", df['Survived'].value_counts())
print("Pclass:\n", df['Pclass'].value_counts())
print("Sex:\n", df['Sex'].value_counts())
```

```python
# Clean data for visualizations
df_cleaned = df[['Survived', 'Age', 'Fare', 'Pclass']].dropna()
```

```python
sns.pairplot(df_cleaned, hue='Survived', palette='coolwarm')
plt.suptitle("Pairplot of Age, Fare, Pclass vs Survived", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

df['Fare'].hist(bins=30, edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survived')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Pclass')
plt.show()
```
